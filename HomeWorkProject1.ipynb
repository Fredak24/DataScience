```python
# Collect CSV files
pattern = "C:/Users/coolr/Google Drive/DataScience_HomeProject/LoanStats_securev1_*.csv"
all_files = glob.glob(pattern)

# Create an empty list
file_List = []

#  Iterate over all_files
for file in all_files:
    df = pd.read_csv(file,  dtype = object)
    file_List.append(df)

# Concatenate frames into a single DataFrame
combined_file = pd.concat(file_List, axis=0)

combined_file.to_csv("combined_file.csv", index=False)
```
